# LFV Higgs Decays in Scotogenic Model

This model is a simple extension Standard Model which includes an extra $Z_2$ symmetry, second Higgs doublet and finally three right handed neutrinos,  all this new fields are odd under $Z_2$ symmetry. 
The matter content of the Scotogenic Model with the group $SU(2)_L\otimes U(1)_Y\otimes Z_2$ is given by [Ma 2006](https://journals.aps.org/prd/abstract/10.1103/PhysRevD.73.077301): 
$$
\left(\nu_{L i}, l_{L i}\right) \sim(2,-1 / 2,+),\quad l_{R i} \sim(1,1,+), \quad N_{R i} \sim(1,0,-)$$
$$
\left(\phi^{+}, \phi^{0}\right) \sim(2,1 / 2,+), \quad\left(\eta^{+}, \eta^{0}\right) \sim(2,1 / 2,-)
$$
Yukawa sector allows the interaction between left and right handed neutrinos including $\eta$ field
$$
\mathcal{L}_{Y}=f_{i j}\left(\phi^{-} \nu_{L i}+\overline{\phi_{0}} l_{L i}\right) l_{R j}+y_{i j}\left(\nu_{L i} \eta_{0}-l_{L i} \eta^{+}\right) N_{R j}+ \textrm{h. c.},
$$
and the Majorana term to right handed neutrinos is added.
$$
\frac{1}{2} M_{i} N_{R i} N_{R i}+h . c .
$$
In this case the more general scalar potential under $Z_2$ symmetry is given by
$$
V(\Phi, \eta)= \mu_{1}^{2}\left(\Phi^{\dagger} \Phi\right)+\mu_{2}^{2}\left(\eta^{\dagger} \eta\right)+\frac{\lambda_{1}}{2}\left(\Phi^{\dagger} \Phi\right)^{2}+\frac{\lambda_{2}}{2}\left(\eta^{\dagger} \eta\right)^{2} +\lambda_{3}\left(\Phi^{\dagger} \Phi\right)\left(\eta^{\dagger} \eta\right)+\lambda_{4}\left(\Phi^{\dagger} \eta\right)\left(\eta^{\dagger} \Phi\right)+\frac{\lambda_{5}}{2}\left[\left(\Phi^{\dagger} \eta\right)^{2}+ \textrm{h. c.} \right].
$$
The mass spectrum to new scalar particles is
$$
m^{2}(\sqrt{2} R e \phi_{0})=m_h^2 =2 \lambda_{1} v^{2},$$ 
$$ m^{2}\eta^{ \pm}=m_{\eta}^2 =\frac{\lambda_{3} v^{2}}{2}+m_{2}^{2},$$ 
$$m^{2}(\sqrt{2} Re \eta_{0})=m_{R}^2 =2 m_{2}^{2}+v^{2}\left(\lambda_{3}+\lambda_{4}+\lambda_{5}\right), $$ 
$$m^{2}\sqrt{2} Im \eta_{0}=m_I^2=2 m_{2}^{2}+v^{2}\left(\lambda_{3}+\lambda_{4}-\lambda_{5}\right),
$$
and SM-like Higgs is associated to $h = \sqrt{2}\text{Re}\phi_0 $.

Neutrino mass is generated by a radiative seesaw mechanism. The analytic expression is given by
$$
\left(\mathbf{\mathcal{M}}_\nu\right)_{i j}= y_{i k} y_{j k} M_{k} \frac{1}{16 \pi^{2}}\left[\left(\frac{m_{R}^{2}}{M_{k}^{2}-m_{R}^{2}} \log \frac{m_{R}^{2}}{M_{k}^{2}}\right)-\left(\frac{m_{I}^{2}}{M_{k}^{2}-m_{I}^{2}} \log \frac{m_{I}^{2}}{M_{k}^{2}}\right)\right],
$$
where $m_R$, $m_I$, $m_\eta$ and $M_k$ are the masses of $\eta_R$, $\eta_I$, $\eta^{\pm}$ and $N_{Ri}$ respectively. Also, $k$ index sums over heavy neutrino masses. In the particular case in which  $m_R^2 - m_I^2 = 2 \lambda_{5} v^2$ is assumed smaller than $m_0^2 = (m_R^2 + m_I^2)/2$ we have the next simplification 

$$
\left(\mathcal{M}_{\nu}\right)_{i j}=\frac{\lambda_{5} v^{2}}{8 \pi^{2}} \sum_{k} \frac{y_{i k} y_{j k} M_{k}}{m_{0}^{2}-M_{k}^{2}}\left[1-\frac{M_{k}^{2}}{m_{0}^{2}-M_{k}^{2}} \ln \frac{m_{0}^{2}}{M_{k}^{2}}\right].
$$
Which can be rewritten in a matrix form as follows
$$
\mathbf{\mathcal{M}}_\nu = \mathbf{Y}\mathbf{D}_{\Lambda}\mathbf{Y}^{\top},
$$
where
$$
\mathbf{D}_{\Lambda} = \text{diag}(\Lambda_1,\Lambda_2,\Lambda_3)$$
$$\Lambda_k = \frac{\lambda_{5} v^2}{16 \pi M_k} I(\frac{M_k^2}{m_0^2}), $$
$$I(x) = \frac{x}{1-x} + \frac{x^2 \ln{x}}{(1-x)^2},
$$
To other side, Yukawa matrix can be parametrized in terms of neutrinos masses and one orthogonal matrix by means of Casas-Ibarra parmetrization [Casas, Ibarra 2001](https://www.sciencedirect.com/science/article/pii/S0550321301004758?via%3Dihub)
$$
\mathbf{Y} = \mathbf{D}_{(\sqrt{\Lambda})^{-1}} \mathbf{R} \mathbf{D}_{\sqrt{m_{\nu}}} \mathbf{U}^{\dagger}
$$
where $\mathbf{U}$ is the Pontercorvo-Maki-Nakata-Sakata matrix, also the next notation is followed for diagonal matrices,
$$
\mathbf{D}_{(\sqrt{\Lambda})^{-1}} = \text{diag}(\sqrt{\Lambda_1}^{-1},\sqrt{\Lambda_2}^{-1},\sqrt{\Lambda_3}^{-1}),\,\,
\mathbf{D}_{(\sqrt{m_\nu})} = \text{diag}(\sqrt{m_{\nu_1}},\sqrt{m_{\nu_2}},\sqrt{m_{\nu_3}}),
$$ 
To other hand, $\mathbf{R}$ orthogonal matrix can be parametrized in general by three complex angles  $\theta_{1,2,3}$ as follows:
$$
\mathbf{R}=\left(\begin{array}{ccc}
{c_{2} c_{3}} & {-c_{1} s_{3}-s_{1} s_{2} c_{3}} & {s_{1} s_{3}-c_{1} s_{2} c_{3}} \\
{c_{2} s_{3}} & {c_{1} c_{3}-s_{1} s_{2} s_{3}} & {-s_{1} c_{3}-c_{1} s_{2} s_{3}} \\
{s_{2}} & {s_{1} c_{2}} & {c_{1} c_{2}}
\end{array}\right)
$$
where the notation $c_i \equiv \cos{\theta_i}$ and $s_i \equiv \sin{\theta_i}$ has been used.
Finally, for LFV Higgs decays the pertinent vertexes in this model are given in  the next table.

|          Vertex          | Coupling                                               |          Vertex          | Coupling|

## Heavy neutrino contributions to LFV Higgs Decays

Scotogenic model has a total split between standard model particles and new particles as a consequence of $Z_2$ symmetry. Then, LFV Higgs decays has separate contributions of light neutrinos ($\nu_i$) and heavy neutrinos ($N_i$). Light neutrino contribution is smaller due to light neutrino mas order ($<1$eV). To other side, heavy neutrino contribution has a wide allowed parameter space where we can find higher values of $Br(h \rightarrow l_i l_j)$, and the number of diagrams to consider is only three. The diagrams involved are only three and only charged inert scalars and heavy neutrinos appear.

```python
import OneLoopLFVHD as lfvhd
from sympy import symbols, init_printing, conjugate,I,IndexedBase,Matrix,Add,MatrixSymbol,sqrt
init_printing()
```

**Defining symbolic variables**

```python
μ2,v = symbols(r'\mu_2,v',real=True)
Yν = IndexedBase(r'{{Y^\nu}}')
l,i,j = symbols('l,i,j',integer=True)

## Masses of heavy neutrino l and charged scalars
mNul,mη = symbols(r'm_{{N_l}},m_{{\eta}}',positive=True)
masasNηη = [mNul,mη,mη]

λ3 = (2/v**2)*(mη**2-μ2**2)
```

**Defining Vertexes**

```python
vertex_hηuηd = lfvhd.VertexHSS(-I*λ3*v)
vertex_ηdljNl = lfvhd.VertexSFF(0,I*conjugate(Yν[l,j]))
vertex_ηuliNl = lfvhd.VertexSFF(I*Yν[l,i],0)
```

### Triangle diagram $N_l \eta^+ \eta^-$

```python
TriangleNlηuηd = lfvhd.TriangleFSS(vertex_hηuηd,vertex_ηdljNl,vertex_ηuliNl,masasNηη)
```

```python
TriangleNlηuηd.AL().subs(lfvhd.cambiosDivFin(mNul,mη,mη))#.subs(lfvhd.cambios_aprox(mNul,mη,mη))
```

```python
TriangleNlηuηd.AL()#_GIM()#.factor()
```

```python
TriangleNlηuηd.AR()#_GIM().simplify()
```

### Bubble $N_l \eta$

```python
g2,mW = symbols('g,m_W',positive=True)
c = lambda m:(I*g2*m)/(2*mW)
```

```python
vertex_hljlj = lfvhd.VertexHFF((I*g2*lfvhd.mj)/(2*mW))
BubbleNlη = lfvhd.BubbleFS(vertex_hljlj,vertex_ηdljNl,vertex_ηuliNl,[mNul,mη])
BubbleNlη
```

```python
BubbleNlη.AL().subs(lfvhd.cambiosDivFin(mNul,mη,mη)).simplify()
```

```python
BubbleNlη.AR().subs(lfvhd.cambiosDivFin(mNul,mη,mη)).simplify()
```

### Bubble $\eta N_l$

```python
vertex_hlili = lfvhd.VertexHFF((I*g2*lfvhd.mi)/(2*mW))
BubbleηNl = lfvhd.BubbleSF(vertex_hlili,vertex_ηdljNl,vertex_ηuliNl,[mNul,mη])
BubbleηNl
```

```python
BubbleηNl.AL().subs(lfvhd.cambiosDivFin(mNul,mη,mη)).simplify()
```

## Form factor in this base

```python
from sympy import Symbol,Add,pi,log,lambdify
```

```python
ALN#.atoms(Symbol)
```

```python
ml = IndexedBase('m_l',positive=True)
M = IndexedBase('M',positive=True)
ALNij = lambda ii,jj:Add(*[ALN.subs({lfvhd.mi:ml[i],
                                     lfvhd.mj:ml[j],mNul:M[l]}).subs({i:ii,j:jj,l:k}) 
              for k in range(1,4)])
ARNij = lambda ii,jj:Add(*[ARN.subs({lfvhd.mi:ml[i],
                                     lfvhd.mj:ml[j],mNul:M[l]}).subs({i:ii,j:jj,l:k}) 
              for k in range(1,4)])
```

```python
ALNij(3,2)
```

```python
cambiosYs = {Yν[i,j]:YMf[i-1,j-1] for i in range(1,4) for j in range(1,4)}
cambiosYs
```

```python
conts = {mh:125.18,v:246}
Δ21,Δ31 = symbols(r'\Delta_{21},\Delta_{31}',positive=True)
ms = {m2:sqrt(m1**2 + Δ21),m3:sqrt(m1**2 + Δ31)}
num = {Δ21:7.39e-23,#GeV^2
       Δ31:2.525e-21#GeV^2
      }

M1,M2,M3 = symbols('M_1,M_2,M_3',positive=True)
Ms = {M[1]:M1,M[2]:M2,M[3]:M3}

mls = {ml[3]:1.77686,ml[2]:0.10566}#GeV

def Deltak(mR,mI,mk):
    return ((mk)/(32*pi**2)*(mR**2/(mR**2 - mk**2)*log(mR**2/mk**2) - 
                          mI**2/(mI**2 - mk**2)*log(mI**2/mk**2)))


mR,mI,λ5 = symbols('m_R,m_I,\lambda_5',positive=True)
mIexp2 = mR**2 - 2*λ5*v**2
Λs = {Λ1:Deltak(mR,mI,M1).subs(mI,mIexp2),
            Λ2:Deltak(mR,mI,M2).subs(mI,mIexp2),
            Λ3:Deltak(mR,mI,M3).subs(mI,mIexp2)}

### h-> tau mu
ALNtaumu = ALNij(3,2).subs(cambiosYs).subs(Λs).subs(Ms).subs(
    lfvhd.cambios_aprox(M1,mη,mη)).subs(
    lfvhd.cambios_aprox(M2,mη,mη)).subs(
    lfvhd.cambios_aprox(M3,mη,mη)).subs(angles_osc).subs(ms).subs(num).subs(mls).subs(
    m1,3.01e-11).subs(lfvhd.ma,mh).subs(conts)

ARNtaumu = ARNij(3,2).subs(cambiosYs).subs(Λs).subs(Ms).subs(
    lfvhd.cambios_aprox(M1,mη,mη)).subs(
    lfvhd.cambios_aprox(M2,mη,mη)).subs(
    lfvhd.cambios_aprox(M3,mη,mη)).subs(angles_osc).subs(ms).subs(num).subs(mls).subs(
    m1,3.01e-11).subs(lfvhd.ma,mh).subs(conts)

### h->mu tau

ALNmutau = ALNij(2,3).subs(cambiosYs).subs(Λs).subs(Ms).subs(
    lfvhd.cambios_aprox(M1,mη,mη)).subs(
    lfvhd.cambios_aprox(M2,mη,mη)).subs(
    lfvhd.cambios_aprox(M3,mη,mη)).subs(angles_osc).subs(ms).subs(num).subs(mls).subs(
    m1,3.01e-11).subs(lfvhd.ma,mh).subs(conts)

ARNmutau = ARNij(2,3).subs(cambiosYs).subs(Λs).subs(Ms).subs(
    lfvhd.cambios_aprox(M1,mη,mη)).subs(
    lfvhd.cambios_aprox(M2,mη,mη)).subs(
    lfvhd.cambios_aprox(M3,mη,mη)).subs(angles_osc).subs(ms).subs(num).subs(mls).subs(
    m1,3.01e-11).subs(lfvhd.ma,mh).subs(conts)
```

```python
ALNtaumu.atoms(Symbol),ARNmutau.atoms(Symbol)
```

```python
from scipy.special import spence
def sci_polylog(s,z):
    return spence(1-z)
```

```python
### h-> tau mu
ALNtaumunp = lambdify([M1,M2,M3,mR,mη,μ2,λ5],ALNtaumu,modules=['numpy',{'polylog':sci_polylog}])
ARNtaumunp = lambdify([M1,M2,M3,mR,mη,μ2,λ5],ARNtaumu,modules=['numpy',{'polylog':sci_polylog}])
```

```python
### h-> mu tau

ALNmutaunp = lambdify([M1,M2,M3,mR,mη,μ2,λ5],ALNmutau,modules=['numpy',{'polylog':sci_polylog}])
ARNmutaunp = lambdify([M1,M2,M3,mR,mη,μ2,λ5],ARNmutau,modules=['numpy',{'polylog':sci_polylog}])
```

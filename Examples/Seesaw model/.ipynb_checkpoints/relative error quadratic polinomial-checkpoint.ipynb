```python
UnuOne = {mn[i]:mni,Uν[b,i]:Ubi,Uνc[a,i]:Ucai}
def sumOneS(a,b):
    ALOne = 0
    AROne = 0
    for diag in TrianglesOneFermion:
        ALOne += diag.AL().subs(lfvhd.D,4).subs(
            cambios_hab(a,b)).subs(valores).subs(UnuOne)
        AROne += diag.AR().subs(lfvhd.D,4).subs(
            cambios_hab(a,b)).subs(valores).subs(UnuOne)
    for diag in Bubbles:
        ALOne += diag.AL().subs(lfvhd.D,4).subs(
            cambios_hab(a,b)).subs(valores).subs(UnuOne)
        AROne += diag.AR().subs(lfvhd.D,4).subs(
            cambios_hab(a,b)).subs(valores).subs(UnuOne)
    return ALOne.simplify(),AROne.simplify()
```

```python
ALOne23,AROne23 = sumOneS(2,3)
ALOne32,AROne32 = sumOneS(3,2)

ALOne13,AROne13 = sumOneS(1,3)
ALOne31,AROne31 = sumOneS(3,1)

ALOne12,AROne12 = sumOneS(1,2)
ALOne21,AROne21 = sumOneS(2,1)
```

```python
def GIM_One(exp):
    args = exp.expand().args
    X = Add(*[t for t in args if t.has(mni)])
    #X1 = X.collect([mni],evaluate=False)
    return X.simplify()#mni**2*X1[mni**2]
```

```python
def sumTwoS(a,b):
    ALTwo = 0
    ARTwo = 0
    #UnuTwo = {mn[i]:mni,mn[j]:mnj,C[i,j]:Cijs, Cc[i,j]:Cijcs, Uν[b,j]:Ubj, Uνc[a,i]:Ucai}
    for diag in TrianglesTwoFermion:
        ALTwo += diag.AL().subs(lfvhd.D,4).subs(
            cambios_hab(a,b)).subs(valores).subs(UnuTwo)
        ARTwo += diag.AR().subs(lfvhd.D,4).subs(
            cambios_hab(a,b)).subs(valores).subs(UnuTwo)
    return ALTwo.simplify(),ARTwo.simplify()
```

```python
def sumTwoL(m6,ALab,a,b):
    AL = 0  
    mnk = [m1,m2,m3,m4(m6),m5(m6),m6]
    for i in range(1,7):
        for j in range(1,7):
            L = ALab(mnk[i-1],mnk[j-i],Unu(mnk[0],mnk[-1],b,j),np.conjugate(Unu(mnk[0],mnk[-1],a,i)),
                      Cij(mnk[0],mnk[-1],i,j),np.conjugate(Cij(mnk[0],mnk[-1],i,j)))
            #print(L)
            AL += L#np.nan_to_num(L)
    return AL

def sumTwoR(m6,ARab,a,b):
    AR = 0  
    mnk = [m1,m2,m3,m4(m6),m5(m6),m6]
    for i in range(1,7):
        for j in range(1,7):
            R = ARab(mnk[i-1],mnk[j-i],Unu(mnk[0],mnk[-1],b,j),np.conjugate(Unu(mnk[0],mnk[-1],a,i)),
                      Cij(mnk[0],mnk[-1],i,j),np.conjugate(Cij(mnk[0],mnk[-1],i,j)))
            AR += R#np.nan_to_num(R)
    return AR
```

```python
ALN = -I*g/(2*mW)*(mni*Cijs + mnj*Cijcs)
ARN = -I*g/(2*mW)*(mnj*Cijs + mni*Cijcs)
ALN,ARN
```

```python
ALNnp = lambdify([mni,mnj,Cijs,Cijcs],ALN.subs(valores),'numpy')
ARNnp = lambdify([mni,mnj,Cijs,Cijcs],ARN.subs(valores),'numpy')
def ΓhNiNj(AL,AR,ma=valores[mh],mi=1,mj=60):
    return np.where(np.real(mi+mj)/2<ma/2,Γhlilj(AL,AR,ma=ma,mi=mi,mj=mj),0)
#ΓhNiNj = np.vectorize(ΓhNiNj)
```

```python
X = lambda AL,AR,a,b:Γhlilj(AL,AR,ma=valores[mh],mi=ml[a],mj=ml[b])
sumatot = lambda m6: (X(ALtot23(m6),ARtot23(m6),2,3) + X(ALtot13(m6),ARtot13(m6),1,3)+ X(ALtot12(m6),ARtot12(m6),1,2)
                     + X(ALtot32(m6),ARtot32(m6),3,2) + X(ALtot31(m6),ARtot31(m6),3,1)+ X(ALtot21(m6),ARtot21(m6),2,1))

def sumaNtot(m6,ma =valores[mh]):
    
    mN = {4:m6/3,5:m6/2,6:m6}
    sumaN = sum(np.nan_to_num(ΓhNiNj(ALNnp(mN[i],mN[j],Cij(m1,m6,i,j),np.conjugate(Cij(m1,m6,i,j))),
                          ARNnp(mN[i],mN[j],Cij(m1,m6,i,j),np.conjugate(Cij(m1,m6,i,j))),
                         ma=ma,mi=mN[i],mj=mN[j]))
                   for i in [4,5,6] for j in [4,5,6] if i>=j)
    return sumaN

def BRhl23(m6,ma=valores[mh]):
    suma = sumatot(m6)
    sumaN = sumaNtot(m6)
    return np.nan_to_num(Γhlilj(ALtot23(m6),ARtot23(m6),ma=ma,mi=ml[2],mj=ml[3]))/(suma + 0.0032+ sumaN )

def BRhl13(m6,ma=valores[mh]):
    suma = sumatot(m6)
    sumaN = sumaNtot(m6)
    return np.nan_to_num(Γhlilj(ALtot13(m6),ARtot13(m6),ma=ma,mi=ml[1],mj=ml[3]))/(suma + 0.0032 + sumaN )

def BRhl12(m6,ma=valores[mh]):    
    suma = sumatot(m6)
    sumaN = sumaNtot(m6)
    return np.nan_to_num(Γhlilj(ALtot12(m6),ARtot12(m6),ma=ma,mi=ml[1],mj=ml[2]))/(suma + 0.0032 + sumaN )
```

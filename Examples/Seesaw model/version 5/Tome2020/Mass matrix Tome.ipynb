# LFV Higgs decays in SeeSaw model

Study of neutrino mass matrix in the work  [$\rho$ parameter and $H \to \ell_i \ell_j$ in models with TeV sterile neutrinos](https://doi.org/10.1103/PhysRevD.102.113006)

## Define symbolic variables

Importing sympy functions

```python
from sympy import symbols, init_printing, simplify,factor, Matrix, solve, Eq
init_printing()
```

```python
m1, m2, m3, M, mu = symbols(r'm1, m2, m3, M, \mu',positive=True)
Mnu = Matrix([
    [0, 0, 0, 0, m1],
    [0, 0, 0, 0, m2],
    [0, 0, 0, 0, m3],
    [0, 0, 0, 0, M],
    [m1, m2, m3, M, mu]
    ])
Mnu
```

```python
Mnu.eigenvals()
```

```python
eigv = Mnu.eigenvects()
eigv[0]
```

```python
eigv[1]
```

```python
eigv[2]#[2]
```

```python
e1 = eigv[0][2][0].T
e2 = eigv[0][2][1].T
e3 = eigv[0][2][2].T
e4 = eigv[1][2][0].T
e5 = eigv[2][2][0].T
R = Matrix([e1,e2, e3,e4,e5])
```

```python
simplify((e1.T*e4)[0])
```

```python
simplify(sum([R[3-1,k]*R.T[1-1,k] for k in range(5)])).subs(mu,0)
```

```python
dia = simplify(R*Mnu*R.T)
```

```python
dia
```

```python
Mn1,Mn2,mn = list(Mnu.eigenvals().keys())
Mn1,Mn2
```

```python
(Mn1 + Mn2).expand()
```

```python
(Mn2 - Mn1).expand()
```

```python
simplify(Mn2**2/Mn1**2)
```

```python
Mn1sp, Mn2sp = symbols('M_{n_1}, M_{n_2}',positive=True)
```

```python
f = ((Mn2 - Mn1)**2).subs(mu**2,(Mn1sp+Mn2sp)**2)
f
```

```python
ec1 = Eq(f,(Mn2sp - Mn1sp)**2)
ec1
```

```python

solve(ec1,M**2)[0]
```

```python

Mnu2 = Mnu.subs({mu:Mn1sp + Mn2sp, M:-(m1**2 + m2**2 + m3**2) - Mn1sp*Mn2sp})
Mnu2
```

```python
v1,v2,v3 = list(Mnu2.eigenvals().keys())
```

```python
simplify(v1.subs(Mn2sp,Mn1sp))
```

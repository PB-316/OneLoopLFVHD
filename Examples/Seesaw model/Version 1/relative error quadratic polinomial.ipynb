## In this notebook we use the numpy implementations solutions $x_1$, $x_2$ and $y_{ij}$

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from OneLoopLFVHD.LFVHDFeynG_numpy import rootsx2_inv, x1,x2, y11, y12,y11np,y12np, M0,M1,_ma,_mi
from OneLoopLFVHD.roots import rootsx2
import numpy as np
```

```python
mh = 125.10
a = mh**2
mi = 0.10566
m1 = 80.379
m0 = np.logspace(-12,15,100)
b = - (mi**2 + m1**2-m0**2)
c = m1**2
# y11num, y12num = rootsx2_inv(a,b,c)
y11num, y12num = rootsx2(a,b,c)
#y11num, y12num 
```

```python
def er(y,yaprox):
    return np.abs(y-yaprox)/np.abs(y)
```

```python
plt.loglog(m0,y11num,label='$y_{11}^{num}$')
plt.loglog(m0,y12num,label='$y_{12}^{num}$')
plt.legend()
```

```python
plt.loglog(m0,y11np(mh,mi,m0,m1),label='$y_{11}^{np}$')
plt.loglog(m0,y12np(mh,mi,m0,m1),label='$y_{12}^{np}$')
plt.legend()
```

```python
plt.loglog(m0,er(y11num,y11np(mh,mi,m0,m1)),label='$e_{11}^{np}$')
plt.loglog(m0,er(y12num,y12np(mh,mi,m0,m1)),label='$e_{12}^{np}$')
plt.legend()
```

```python
y11
```

```python
y11sym = np.array([y11.evalf(subs={_ma:mh,_mi:mi,M1:m1,M0:m}) for m in m0],dtype=np.complex256)
y12sym = np.array([y12.evalf(subs={_ma:mh,_mi:mi,M1:m1,M0:m}) for m in m0],dtype=np.complex256)
```

```python
plt.figure(figsize=(12,8))
plt.loglog(m0,er(y11sym,y11num),label='$e_{11}^{sym-num}$')
plt.loglog(m0,er(y12sym,y12num),label='$e_{12}^{sym-num}$')

plt.xlabel('$M_0$',fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.title(r'$m_i = m_\mu$, $m_a = m_h$, $M_1 = m_W$',fontsize=18)
#np.abs(y-yaprox)/np.abs(y)
error = r'$e_r = |\frac{y_{sym} - y_{num}}{y_{sym}}|$';
y_sol = r'$y_{1,2} = \frac{-b \mp \sqrt{b^2 - 4 a c}}{2 a}$';

plt.text(1e-10,1e21,error,fontsize=24)
plt.text(1e-10,1e7,y_sol,fontsize=24)
plt.legend(fontsize=18,frameon=False)
plt.savefig('ys_inestable.png')
```

```python
plt.loglog(m0,er(y11sym,y11np(mh,mi,m0,m1)),label='$e_{11}^{sym-np}$')
plt.loglog(m0,er(y12sym,y12np(mh,mi,m0,m1)),label='$e_{12}^{sym-np}$')
plt.legend(fontsize=15)
```

```python
plt.loglog(m0,er(y11sym,y11np(mh,mi,m0,m1)),label='$e_{11}^{sym-np}$')
plt.loglog(m0,er(y12sym,y12num),label='$e_{12}^{sym-num}$')
plt.legend(fontsize=15)
```

```python
convert_to_array = lambda df,col:np.array(list(map(mpmathify,df[col])))

def mpmathify_frame(df):
    dict = {}
    for col in df:
        #print(col)
        dict[col] = convert_to_array(df,col)
        #print(dict.keys())
    #print(dict.keys())
    return pd.DataFrame(dict)
```

```python
cols = ['niWW','niWG','niGW','niGG','niW','niG','Wni','Gni','Gninj','Wninj']
colstwo = ['Gninj','Wninj']
def total_frame(FF_df,nf=1):
    if nf == 1:
        FF_df['total'] = FF_df[cols].sum(axis=1)
    elif nf == 2:
        FF_df['total'] = FF_df[colstwo].sum(axis=1)
    else:
        raise ValueError('nf must be 1 or 2')
```

```python
diagG = ['Gninj','niGG','niG','Gni']
diagW = ['Wninj','niWW','niW','Wni']#'niWG','niGW'
diagWG = ['niWG','niGW']
def suma_WG(frameL,frameR):
    sumaGL = 0
    sumaGR = 0
    sumaWL = 0
    sumaWR = 0
    sumaWGL = 0
    sumaWGR = 0
    for dia in diagG:
        sumaGL += frameL[dia]
        sumaGR += frameR[dia]
    for dia in diagW:
        sumaWL += frameL[dia]
        sumaWR += frameR[dia]
    for dia in diagWG:
        sumaWGL += frameL[dia]
        sumaWGR += frameR[dia]
    return {'G':(sumaGL, sumaGR), 'W':(sumaWL, sumaWR), 'WG':(sumaWGL, sumaWGR)}
```

```python
suma_total = suma_WG(FF_frame23L_total,FF_frame23R_total)
```
